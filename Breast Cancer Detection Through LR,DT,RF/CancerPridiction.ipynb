```python
X = df.iloc[:,2:31].values
Y = df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
X_tr, X_te, Y_tr, Y_te = train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
X_tr = StandardScaler().fit_transform(X_tr)
X_te = StandardScaler().fit_transform(X_te)
```

```python
def models(X_tr, Y_tr):
    log=LogisticRegression(random_state=0)
    log.fit(X_tr, Y_tr)
    
    tree = DecisionTreeClassifier(random_state=0, criterion="entropy")
    tree.fit(X_tr, Y_tr)
    
    forest=RandomForestClassifier(random_state=0,criterion="entropy",n_estimators=10)
    forest.fit(X_tr, Y_tr)
    
    print("LogisticRegression: ", log.score(X_tr, Y_tr))
    print("DecisionTreeClassifier: ", tree.score(X_tr, Y_tr))
    print("RandomForestClassifier: ", forest.score(X_tr, Y_tr))
    
    return log,tree,forest
```

```python
model = models(X_tr,Y_tr)
```

```python
for i in range(len(model)):
    print("Model",i)
    print(classification_report(Y_te,model[i].predict(X_te)))
    print('Accuracy : ',accuracy_score(Y_te,model[i].predict(X_te)))
```
